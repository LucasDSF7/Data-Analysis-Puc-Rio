# Iris dataset

First exemple to practice data exploratory using iris dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
from random import randint
import seaborn as sns
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
```

# Variable Table

Features: sepal length, sepal width, petal length, petal width

Target: class

```python
df = pd.read_csv(url, names=('sepal length', 'sepal width', 'petal length', 'petal width', 'class'))
df.head()
```

```python
df.describe()
```

```python
colors = {'Iris-setosa': 'b', 'Iris-versicolor': 'r', 'Iris-virginica': 'g'}

for i in range(df.shape[0]):
    plt.scatter(df['petal length'].iloc[i], df['petal width'].iloc[i], color=colors[df['class'].iloc[i]])    

plt.title('Iris: petal length x width')
plt.xlabel('Petal length')
plt.ylabel('Petal width')
```

```python
markers = {'Iris-setosa': '.', 'Iris-versicolor': '+', 'Iris-virginica': 'x'}

for i in range(df.shape[0]):
    plt.scatter(df['petal length'].iloc[i], df['petal width'].iloc[i], marker=markers[df['class'].iloc[i]])    

plt.title('Iris: petal length x width')
plt.xlabel('Petal length')
plt.ylabel('Petal width')
```

Let's assume that Iris records was collected ir time order

```python
plt.close()
attributes = df.columns.drop('class')
x = range(df.shape[0])
for attribute in attributes:
    plt.plot(x, df[attribute], label=attribute)
plt.title('Attribute variation in time')
plt.legend()
```

```python
plt.close()
plt.hist(df['petal length'], 6, color='b', edgecolor='black')
plt.title('Petal length distribution')
plt.ylabel('Frequency')
plt.xlabel('Petal Length')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(8, 8))
ax[0, 0].hist(df['petal length'], bins=7)
ax[0, 0].set_title('Petal length')

ax[0, 1].hist(df['petal width'], bins=7)
ax[0, 1].set_title('Petal width')

ax[1, 0].hist(df['sepal length'], bins=7)
ax[1, 0].set_title('Sepal length')

ax[1, 1].hist(df['sepal width'], bins=7)
ax[1, 1].set_title('Sepal width')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(8, 8))
ax[0, 0].boxplot(df['petal length'])
ax[0, 0].set_title('Petal length')

ax[0, 1].boxplot(df['petal width'])
ax[0, 1].set_title('Petal width')

ax[1, 0].boxplot(df['sepal length'])
ax[1, 0].set_title('Sepal length')

ax[1, 1].boxplot(df['sepal width'])
ax[1, 1].set_title('Sepal width')
```
